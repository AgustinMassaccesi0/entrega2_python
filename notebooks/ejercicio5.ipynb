```python
import sys
sys.path.append('../src')
import costo
```

```python
while True:
    try:
        peso = float(input("Ingrese el peso del paquete (En kg): "))
        if peso <= 0:
            print("Error: el peso debe ser mayor a 0")
        else:
            break
    except ValueError:
        print("Error: ingrese un número válido")


while True:
    zona = input("Ingrese la zona (local, regional, nacional): ").strip().lower()
    if zona in ['local', 'regional', 'nacional']:
        break
    print("Error: zona no válida. Las zonas disponibles son: local, regional, nacional")

print(f"Costo de envío: ${costo.costo_envio(peso, zona)}")
```
