```python
import sys
sys.path.append('../src')
import cifrado
```

```python
texto = input('ingrese un mensaje:')
while True:
    desplazamiento = input("Ingrese el desplazamiento: ").strip()
    
    if desplazamiento.lstrip('-').isdigit():
        desplazamiento = int(desplazamiento)
        break
    else:
        print("Error: debe ingresar un número entero.\n")

mensaje_cifrado = cifrado.cifrado_cesar(texto, desplazamiento)
descifrado = cifrado.cifrado_cesar(mensaje_cifrado, -desplazamiento)

print("Mensaje cifrado:", mensaje_cifrado)
print("Mensaje descifrado:", descifrado)
```
