```python
import random


while True:
    participantes = input('Ingrese los participantes (al menos 3 y separados por coma): ')
    
    lista_p = [p.strip() for p in participantes.split(',') if p.strip() != '']
    lista_lower = [p.lower() for p in lista_p]
    
    if len(lista_p) < 3:
        print('Debe haber al menos 3 participantes.\n')
        continue
    
    if len(lista_lower) != len(set(lista_lower)):
        print('Hay nombres duplicados.\n')
        continue
    
    break


while True:
    aleatorio = random.sample(lista_p, len(lista_p))
    
    
    if all(lista_p[i] != aleatorio[i] for i in range(len(lista_p))):
        break


print('\nSorteo de amigo invisible:')
for i in range(len(lista_p)):
    print(f'{lista_p[i]} → {aleatorio[i]}')
```
